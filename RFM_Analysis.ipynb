```python
# The term 'Omnichannel customers' indicates that customers make purchases from both online
# and offline platforms. We need to create new variables for each customer representing the total number
# of purchases and total spending
df["order_num_total"] = df["order_num_total_ever_online"] + df["order_num_total_ever_offline"]
df["customer_value_total"] = df["customer_value_total_ever_online"] + df["customer_value_total_ever_offline"]
df.info()
```

```python
# we need to convert the types of variables representing dates to 'date'
df = df.apply(lambda col: pd.to_datetime(col) if 'date' in col.name else col)
df.info()
```

```python
# Calculating RFM metrics
df["last_order_date"].max() #2021-05-30
analyse_date = dt.datetime(2021, 6, 1)
rfm = pd.DataFrame()
rfm["customer_id"] = df["master_id"]
rfm["recency"] = (analyse_date - df["last_order_date"]).dt.days
rfm["frequency"] = df["order_num_total"]
rfm["monetary"] = df["customer_value_total"]
rfm.head()
```

```python
# Calculating RFM scores
rfm["recency_score"] = pd.qcut(rfm["recency"], 5, labels=[5, 4, 3, 2, 1])
rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm["monetary_score"] = pd.qcut(rfm["monetary"], 5, labels=[1, 2, 3, 4, 5])
rfm["rf_score"] = rfm["recency_score"].astype(str) + rfm["frequency_score"].astype(str)
rfm.head()
```

```python
# Defining Customer Segments Using RF Scores

seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}

rfm["segment"] = rfm["rf_score"].replace(seg_map, regex=True)
rfm.head()
```

```python
# By using RFM analysis, find the customers with below profiles for given case and
# save the customer IDs as a csv file.

## FLO is introducing a new women's shoe brand. The product prices of this new brand
# are above the general customer preferences. Therefore, there is a desire to specifically
# communicate with the interested profile for the introduction and sales of the brand.
# Customers who are loyal (champions, loyal customers) and those who shop in the women's
# category will be contacted. Save the customer IDs of these customers to a CSV file.

target_customer_ids = rfm[rfm["segment"].isin(["champions", "loyal customer"])]["customer_id"]
cust_ids = df[(df["interested_in_categories_12"].str.contains("KADIN")) & (df["master_id"].isin(target_customer_ids))]["master_id"]
cust_ids.to_csv('New_Brand_Target_Customer_ID.csv')
```
